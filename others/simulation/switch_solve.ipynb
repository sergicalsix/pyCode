```python
import sympy
from IPython.display import display
```

```python
sympy.var('k1_,k2_,k3_,k4_,s0_,g_')
solve = sympy.solve(k1_*s0_ - k2_*g_ + k3_*g_**2 / (k4_**2 + g_**2), g_)
sympy.init_printing()
display(solve)
```
